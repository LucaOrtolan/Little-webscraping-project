```python
from bs4 import BeautifulSoup
import requests
```

```python
melodie = requests.get(url = "https://it.wikipedia.org/wiki/M%C3%A9lodie_passag%C3%A8re")
```

```python
soup = BeautifulSoup(melodie.content, "html.parser")
soup
```

```python
data = soup.ol
data
```

```python
songs_with_tags = data.find_all("i")

songs_with_tags
```

```python
songs = []
for song in range(len(songs_with_tags)):
    s = songs_with_tags[song].string
    songs.append(s)
    
songs
    
```

```python
composers_with_tags = data.find_all("a")

composers_with_tags
```

```python
composers = []

for composer in range(len(composers_with_tags)):
    c = composers_with_tags[composer].string
    composers.append(c)
    
composers
```

```python
strdata = str(data)

strdata
```

```python
divided_data = strdata.split(">")

divided_data

```

```python
lenghts_with_tags = []

for i in range(len(divided_data)):
    if ":" in divided_data[i]:
        lenghts_with_tags.append(divided_data[i])
        
lenghts_with_tags
```

```python
lenghts = []

for i in range(len(lenghts_with_tags)):
    lenght = lenghts_with_tags[i][4:8]
    lenghts.append(lenght)

lenghts
    
    
```

```python
Melodie_Passagere = {"Song" : songs, "Composer" : composers, "Lenght" : lenghts}

```

```python
import pandas as pd

dataframe = pd.DataFrame(Melodie_Passagere)

dataframe

```
